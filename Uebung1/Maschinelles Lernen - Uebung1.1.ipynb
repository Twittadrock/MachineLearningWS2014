### Vorhersage des Hauspreises
AUs Streudiagramm lässt sich eine positive Korrelation zwischen TGT und RM und eine negative Korrelation zwischen TGT und LSTAT ablesen. *’AGE ’,’DIS ’,’RAD ’,’TAX'* könnte weggelassen werden.

```python
corMat = np.corrcoef(boston.transpose())
shape = corMat.shape
threshold = 0.69

for n in range(shape[1]):
    index = cols.index("TGT")
    if np.abs(corMat[index,n]) >= threshold:
        if(n != index):
            print "TGT" + " x " + cols[n] + " = " + str(corMat[index,n])
            boston.plot(x='TGT', y=cols[n], style='o');
        
```
