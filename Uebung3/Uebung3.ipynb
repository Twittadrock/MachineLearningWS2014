```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%install_ext http://raw.github.com/jrjohansson/version_information/master/version_information.py
%load_ext version_information
```

```python
scale=5.0
size=1000

locx1 = 1
locy2 = 2
x1 = np.random.normal(locx1, scale, size)
y1 = np.random.normal(locy2, scale, size)

locx2 = 30
locy2 = 30
x2 = np.random.normal(locx2, scale, size)
y2 = np.random.normal(locy2, scale, size)
```

```python
plt.hist(x1, bins=50, range=None, normed=False, weights=None, cumulative=False, bottom=None, histtype=u'bar', align=u'mid', orientation=u'vertical', rwidth=None, log=False, color=None, label=None, stacked=False, hold=None);
```

```python
plt.hist(x2, bins=50, range=None, normed=False, weights=None, cumulative=False, bottom=None, histtype=u'bar', align=u'mid', orientation=u'vertical', rwidth=None, log=False, color=None, label=None, stacked=False, hold=None);
```

```python
class1 = np.ones(len(x1))
data1 = np.c_[x1, y1, class1]

class2 = class1*-1
data2 = np.c_[x2, y2, class2]

designMatrix = np.concatenate((data1, data2))
designMatrix.shape
```

```python
designMatrix.shape
designMatrixDf = pd.DataFrame(designMatrix)
```

```python
colors = np.where(designMatrixDf[2] == 1, 'r', 'b')
designMatrixDf.plot(kind='scatter', x=0 , y=1, c=colors)
```

Ausführung der BatchPerzeptron. Übergabe der zuvor erstellten Design-Matrix

```python
w, b = perz.batchPerzeptron(designMatrix)
```

```python
fig, axs = plt.subplots(1)

x = np.linspace(-20, 50, 50)
y = np.linspace(-60, 80, 50)
X, Y = np.meshgrid(x,y)
ZX = X * w[0]
ZY = Y * w[1]
Zdata = ZX + ZY + b 

levels = np.linspace(-200, 200, 20)
cs = axs.contourf(X, Y, Zdata, levels=levels)
fig.colorbar(cs, format="%.2f")

plt.show()
```
